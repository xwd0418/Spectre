```python
import torch, os, pickle
from rdkit import Chem
from rdkit.Chem import Draw

count = 0
with open("/workspace/SMILES_dataset/train/SMILES/index.pkl", "rb") as f:
    smiles = pickle.load(f)
    
for f_name in os.listdir("/workspace/SMILES_dataset/train/HSQC/"):
    data = torch.load("/workspace/SMILES_dataset/train/HSQC/"+f_name)
    if count == 100:
        break
    if not (data<0).any():
        # print(f_name, data)
        count += 1
        mol = Chem.MolFromSmiles(smiles[int(f_name.split(".")[0])])
        img = Draw.MolToImage(mol)
        img.show()
        
```
