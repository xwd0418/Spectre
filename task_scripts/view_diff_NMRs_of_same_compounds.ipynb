```python
def print_diff(sorted_NMRs, is_carbon  ):
        C_or_H = "Carbon" if is_carbon else "Proton"
        NMR_range = 180 if is_carbon else 7.5
        print(f"{C_or_H} values are in the range of 0-{NMR_range}")
        
        for i in range(len(sorted_NMRs)):
            for j in range(i+1, len(sorted_NMRs)):
                
                if len(sorted_NMRs[i]) != len(sorted_NMRs[j]):
                    print(f" NMR#{i+1} vs NMR#{j+1}: \n WARNING: different number of peaks\n ")
                    continue
                
                NMR_i = sorted_NMRs[i]
                NMR_j = sorted_NMRs[j]
                diff = torch.abs(NMR_i - NMR_j)
                print(f" NMR#{i+1} vs NMR#{j+1}:\n avg diff: {diff.mean().item():.4f}, total diff: {diff.sum().item():4f}, percentage diff (avg diff divided by the range of displayed values):{100*diff.mean().item()/NMR_range:.4f}% \n")
```

```python
counter = 0
for smiles, nmrs in SMILES_to_NMRS.items():
    if counter > 10:
        break
    if len(nmrs) > 1:
        counter += 1
        print(f"\n\n\nNumber: {counter}")
        print("SMILES: ", smiles)
        print("Name:" , name_2d[int(nmrs[0].split(".")[0])])
        carbon_positions = []
        hydrogen_positions = []
        for nmr in nmrs:
            
            hsqc = torch.load(open(dataset_path/nmr, 'rb'))
            print(f"Number of HSQC peaks: {len(hsqc)}")
            plot_NMR(hsqc)
            
            carbon_positions.append(hsqc[:,0].sort()[0])
            hydrogen_positions.append(hsqc[:,1].sort()[0])
        print_diff(carbon_positions, True  )
        print_diff(hydrogen_positions, False  )
            
```
