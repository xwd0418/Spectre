```python
from rdkit import Chem
from rdkit.Chem import Descriptors

def get_molecular_weight(smiles):
    # Convert the SMILES string to a molecule object
    molecule = Chem.MolFromSmiles(smiles)
    # Calculate the molecular weight
    return Descriptors.MolWt(molecule)

# Example usage
smiles_string = 'CCO'  # Ethanol
mol_weight = get_molecular_weight(smiles_string)
print(f"The molecular weight of the molecule is: {mol_weight}")
```

```python
import os ,pickle, token
from pathlib import Path
root_path = Path("/workspace/OneD_Only_Dataset")
for split in ['train', 'val', 'test']:
    split_path = root_path / split
    smile_path  = split_path / 'SMILES' / 'index.pkl'
    with open(smile_path, 'rb') as f:
        smile_data = pickle.load(f)
    mw_data = dict([(k, get_molecular_weight(v)) for k, v in smile_data.items()])
    
    mw_path = split_path / 'MW' / 'index.pkl'
    os.makedirs(split_path / 'MW', exist_ok=True)
    with open(mw_path, 'wb') as f:
        pickle.dump(mw_data, f)
    print(f"Processed {smile_path}")
```
