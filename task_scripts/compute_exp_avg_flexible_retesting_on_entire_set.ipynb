```python
import os, pickle, sys, torch, numpy as np


table_path = "/root/gurusmart/MorganFP_prediction/reproduce_previous_works/rank_on_entire_set/larger_flexible_models_3072dim"
table_path = "/root/gurusmart/MorganFP_prediction/reproduce_previous_works/rank_on_entire_set_retrieve_on_deepsat_set/larger_flexible_models_3072dim"
# '/root/gurusmart/MorganFP_prediction/reproduce_previous_works/all_2d1d_datasets/entropy_radius_exps_on_2d1d_datasets'

print_keys = [
    "test_mean_rank_1/",
    "test_mean_rank_5/",
    "test_mean_rank_10/",
    # "test_mean_mean_rank",
    "test_mean_cos",
    "test_mean_f1", 
]


```

```python
def load_result(path):
    total_out_dict = {}
    for modle_type in os.listdir(path):
        # print(modle_type)
        with open(os.path.join(path, modle_type), "rb") as f:
            cur_model = pickle.load(f)[0]
            for k,v in cur_model.items():
                if modle_type.split(".")[0] in k:
                    total_out_dict[k] = v
    return total_out_dict


# load_result(table_path+"/r0_r2_FP_trial_1")
```

```python
all_exps = os.listdir(table_path)
all_exps = sorted(all_exps)
# load pickles 
from collections import defaultdict
exp_name_to_key_to_results = defaultdict(lambda: defaultdict(list))
for exp in all_exps:
    # print(exp)
    if 'trial' in exp:
        trial_spelling = "trial"
    elif 'trail' in exp:
        trial_spelling = "trail"
        
    # with open(f'{table_path}/{exp}/test_result.pkl', 'rb') as f:
    #     result = pickle.load(f)[0]
    print(exp)
    result = load_result(f'{table_path}/{exp}')
      

        
    for print_key in print_keys:
        for pickle_key in result:
            
            if print_key in pickle_key:
                exp_name_to_key_to_results["_".join(exp.split(trial_spelling)[:-1])][pickle_key].append(result[pickle_key])
        # exp_name_to_key_to_results[exp][key].append(result[key])
        # print(key, result[key])
    
            
    
exp_name_to_key_to_results
```

```python
avg_results = {}
for exp_name, key_to_results in exp_name_to_key_to_results.items():
    avg_results[exp_name] = {key: (np.mean(val), np.std(val)) for key, val in key_to_results.items()}
avg_results
```
