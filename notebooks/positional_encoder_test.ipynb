```python
import os, torch
os.chdir("..")
import encoder as p
from matplotlib import pyplot as plt
```

```python
ce = p.CoordinateEncoder((30,30,30, 30,30), 150, wavelength_bounds=[(None, 100)]*5)
t = torch.arange(100).expand(5,100).transpose(0,1).unsqueeze(0)
result = ce(t).squeeze(0)
```

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111)
ax.set_xlabel("input dimension")
ax.set_ylabel("encoded value")
ax.imshow(result, cmap='hot')
fig.show()
```
