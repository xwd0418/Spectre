```python
import os, yaml, glob
from pathlib import Path
```

```python
exp_dir = "/data/smart4.5/new_split"
exp_path = Path(exp_dir)
glob_res = glob.glob(f"{exp_dir}/*/results.yml")
```

```python
first_result = glob_res[0]
with open(first_result, "r") as f:
  first_obj = yaml.safe_load(f)
print("=== Tracked Metrics ===")
for k in first_obj:
  print(k)
```

```python
metrics = ["val/mean_cos", "val/mean_f1", "val/mean_rank_1", "val/mean_rank_5", "val/mean_rank_10"]
import pandas as pd

df_values = []

for glob_path in glob_res:
  with open(glob_path, "r") as f:
    obj = yaml.safe_load(f)
  obj2 = {
    k : obj[k]["maximum"]["value"] for k in metrics
  }
  path = Path(glob_path)
  # crop off results.yml and /data/smart4.5
  experiment_directory = str(path.parent)[len(str(path.parent.parent.parent)):]
  # bulid a data frame
  v = [experiment_directory]
  for k in metrics:
    v.append(obj2[k])
  df_values.append(v)
  # print(f"=== Results path: {glob_path} ===")
```

```python
df = pd.DataFrame(df_values, columns = ["path"] + metrics)
df = df.set_index("path")
df
```
