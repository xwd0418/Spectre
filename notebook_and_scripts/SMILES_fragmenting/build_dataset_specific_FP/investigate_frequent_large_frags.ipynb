```python
from datasets.dataset_utils import  fp_loader_configer

fp_loader_configer.select_version("DB_Specific")
fp_loader = fp_loader_configer.fp_loader
```

```python
entropy_each_frag, counts, total_dataset_size = fp_loader.setup(out_dim=6144, max_radius=10)
frags_to_keep = fp_loader.frag_to_index_map.keys()
```

```python
large_frags = [str(frag) for frag in frags_to_keep if fp_loader.radius_mapping[frag] >= 10]
```

```python
import pickle, torch
chemical_name_path = "/workspace/SMILES_dataset/test/Chemical/index.pkl"
with open(chemical_name_path, 'rb') as f:
    chemical_names = pickle.load(f)
    
SMILES_name_path = "/workspace/SMILES_dataset/test/SMILES/index.pkl"
with open(SMILES_name_path, 'rb') as f:
    SMILES = pickle.load(f)
    
repo_path = "/root/MorganFP_prediction/reproduce_previous_works/Spectre"
split = "test"
path_to_load_full_info_indices = f"{repo_path}/datasets/{split}_indices_of_full_info_NMRs.pkl"

large_frags = set(large_frags)
with open(path_to_load_full_info_indices, 'rb') as f:
    full_info_indices = pickle.load(f)
    
compound_number = 0
for index_f in full_info_indices:
    index = int(index_f.split(".")[0])
    frags = torch.load(f"/workspace/SMILES_dataset/test/fragments_of_different_radii/{index_f}", weights_only=False)
    intersection = large_frags.intersection(frags)
    if intersection:
        compound_number += 1
        print("compound number is", compound_number)
        print("SMILES is", SMILES[index])
        print("name is" , chemical_names[index])
        print("high entropy large frags are", intersection)
        print("\n")
```
